```python
with open("dayFive.txt", "r") as f:
    txt_list = [line.strip() for line in f]
```

```python
data = [i.split(" -> ") for i in txt_list]
data = [[list(map(int,i.split(','))) for i in coord] for coord in data]
sign = lambda a: (a>0) - (a<0)
```

```python
traveledPoints = {}
finalCount = 0
for points in data:
    temp=[]
    cord1 = points[0]
    cord2 = points[1]
    if cord1[0] == cord2[0]:
        tempSign = sign(cord1[1]-cord2[1])*-1
        temp=[[cord1[0],i] for i in range(cord1[1],cord2[1]+tempSign,tempSign)]
    elif cord1[1] == cord2[1]:
        tempSign = sign(cord1[0]-cord2[0])*-1
        temp=[[i,cord1[1]] for i in range(cord1[0],cord2[0]+tempSign,tempSign)]
    for i in temp:
        tempKey = str(i)
        if tempKey in traveledPoints:
            traveledPoints[tempKey] += 1
        else:
            traveledPoints[tempKey] = 1

for key, value in traveledPoints.items():
    if value>1:
        finalCount +=1
print(finalCount)
```
